# Задание 4.2

```python
df = pd.read_csv('precious_metal.csv', sep=';') 
print(df.info())
df.set_axis(['gold', 'silver', 'platinum', 'palladium', 'date'], axis = 'columns', inplace = True) 
print(df.info())
```

# Задание 4.3

```python
df.isnull().sum() 
df.isna().sum() 
df = df.fillna(0)
df.tail(10)
```

# Задание 4.4

```python
def replace_symbol(x): 
 x = str(x) 
 x = float(x.replace(',','.')) 
 return x 
df['gold'] = df['gold'].apply(replace_symbol) 
df
```

# Задание 4.5

```python
sred_mea=df.mean()
sred_mea
```

```python
gold_mean = sred_mea[0]
silver_mean = sred_mea[1]
platinum_mean = sred_mea[2]
palladium_mean = sred_mea[3]
```

```python

df['gold'] = df['gold'].apply(lambda x: gold_mean if x==0 else x)
df['silver'] = df['silver'].apply(lambda x: silver_mean if x==0 else x)
df['platinum'] = df['platinum'].apply(lambda x: platinum_mean if x==0 else x)
df['palladium'] = df['palladium'].apply(lambda x: palladium_mean if x==0 else x)
df.head()
```

# Задание 4.6

```python
df['gold_usd'] = df['gold']/77
df['silver_usd'] = df['silver']/77
df['platinum_usd'] = df['platinum']/77
df['palladium_usd'] = df['palladium']/77
df['gold_eu'] = df['gold']/83
df['silver_eu'] = df['silver']/83
df['platinum_eu'] = df['platinum']/83
df['palladium_eu'] = df['palladium']/83
df
```
