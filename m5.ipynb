```python
def change_month(x):
    if x==4:
        return 'April'
    elif x==5:
        return 'May'
    elif x==6:
        return 'June'
    elif x==7:
        return 'Jule'
    elif x==8:
        return 'AG'
    elif x==9:
        return 'st'
    elif x==10:
        return 'okt'
    else:
        return x


df['month_new'] = df['month'].apply(change_month)
df['month_new']
df
```

```python
import pandas as pd 
df = pd.read_csv('precious_metal_new.csv') 
df = df.loc[:,['gold', 'gold_usd', 'silver', 'silver_usd', 
 'platinum', 'platinum_usd', 'palladium', 'palladium_usd', 
 'date']]
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y %H:%M') 

df['month']=pd.DatetimeIndex(df['date']).month
months = {1:'Январь', 2:'Февраль', 3:'Март', 4:'Апрель', 5:'Май', 6:'Июнь', 7:'Июль', 8:'Август', 9:'Сентябрь', 10:'Октябрь', 11:'Ноябрь', 12:'Декабрь'}
df['month'] = df['month'].apply(lambda i: months[i])
#print(df)

#__________________________________________________________
fn = df[df["month"] == 'Май']
print(fn)
```
