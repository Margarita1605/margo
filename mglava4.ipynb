##  задание 1.4

```python
import matplotlib.pyplot as plt  
import pandas as pd  
df = pd.read_csv('ekb_weather.csv')  
 
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y %H:%M')  
 
 
 
df['month'] = pd.DatetimeIndex(df['date']).month 
 
months = {1:'Январь', 2:'Февраль', 3:'Март', 4:'Апрель', 5:'Май', 6:'Июнь', 7:'Июль', 8:'Август', 9:'Сентябрь', 10:'Октябрь', 11:'Ноябрь', 12:'Декабрь'} 
df['month'] = df['month'].apply(lambda i: months[i]) 
df.to_csv('new_precious_metal.csv') 
print(df) 
print('Сред темп', df['temp'].mean()) 
print('Сред давл',df['pressure'].mean()) 
print('Сред влаж',df['humidity'].mean()) 
 
df_new1=df[df['month']=='Январь'] 
df_new2=df[df['month']=='Декабрь'] 
print(df_new1['temp'].mean()) 
print(df_new2['temp'].mean()) 
 
if (df_new1['temp'].mean()) < (df_new2['temp'].mean()): 
    print ("Январь") 
else: 
    print('Декабрь')
```

```python
df_neew11=df[df['month']=='Апрель'] 
df_neew22=df[df['month']=='Ноябрь'] 
print(df_neew11['pressure'].mean()) 
print(df_neew22['pressure'].mean()) 
 
if (df_neew11['pressure'].mean()) < (df_neew22['pressure'].mean()): 
    print ("Апрель") 
else: 
    print('Ноябрь')
```

```python
df_neew13=df[df['month']=='Июнь'] 
df_neew23=df[df['month']=='Август'] 
print(df_neew13['humidity'].mean()) 
print(df_neew22['humidity'].mean()) 
 
if (df_neew13['humidity'].mean()) < (df_neew23['humidity'].mean()): 
    print ("Июнь") 
else: 
    print('Август')
```
